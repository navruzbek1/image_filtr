## Rasmlarni filtrlash matritsalari orqali tiniqlashtirish

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
image = imread('dog.jpg', as_gray=True)
plt.imshow(image, cmap='gray')
plt.show
```

```python
arr2 = np.array([
    [1,0,1],
    [0,1,0],
    [0,-1,0]
])
arr2
```

```python
a,b = image.shape
c,d = arr2.shape
narr = np.zeros(((a-(c-1)),(b-(d-1))))
for i in range(a-(c-1)):
    for j in range(b-(d-1)):
        narr[i][j] = np.sum(image[i:i+c,j:j+d]*arr2)
narr
```

```python
plt.imshow(image, cmap='gray')
plt.show
```
